Наблюдаем, что в столбце `children` скачок значений с -1 до 20. <p>Общее их кол-во составляет 123 значений. Изменим -1 на 1, а от 20 избавимся. <p>

```python
data['children'] = abs(data['children']) #применяем модуль, чтобы сделать все значения положительными
data = data.loc[data['children'] != 20].reset_index(drop = True) # избавляемся от числа 20
data['children'].value_counts()
```

Таким образом мы сбалансировали значения

```python
positive_days_employed = data['days_employed'][data['days_employed'] >= 0].count()
print('Кол-во положительных значений: ', positive_days_employed, 'дней, что составляет: ', int(positive_days_employed/365), 'лет')
negative_days_employed = data['days_employed'][data['days_employed'] < 0].count()
print('Кол-во отрицательных значений: ', negative_days_employed, 'дней, что составляет: ', int(negative_days_employed/365), 'года')
positive_days_employed_mean = int(data['days_employed'][data['days_employed'] >= 0].mean())
print('Кол-во средних положительных значений: ', positive_days_employed_mean, 'дней, что составляет: ', int(positive_days_employed_mean/365), 'лет')
dob_years_mean = int(data['dob_years'][data['days_employed'] >= 0].median())
print('Медианный возраст: ', dob_years_mean)
negative_days_employed_mean = int(data['days_employed'][data['days_employed'] < 0].mean())
print('Кол-во средних отрицательных значений: ', negative_days_employed_mean, 'дней, что составляет: ', int(negative_days_employed_mean/365), 'лет')
min_positive_days_employed = int(data['days_employed'][data['days_employed'] >= 0].min())
print('Кол-во положительных значений: ', min_positive_days_employed, 'дней, что составляет: ', int(min_positive_days_employed/365), 'лет')
```

**Выводы**
1. Проведя анализ, следует, что отрицательные значения в колонке `days_employed` показыают кол-во трудоыого стажа. Их нужно исправить на положителеьные.
2. Посчитав все положительные значения в колонке `days_employed` можно предположить, что, скорее всего это не дни, а часы, потому что, даже самое минимальное значение привышает средний возраст заёмщика.

```python
def positive(days):
    if days > 0:
        days = days / 24 / 365
    return days

data['days_employed'] = data['days_employed'].apply(positive) # переводим все положительные значения в дни
data['days_employed'] = abs(data['days_employed']) # применяем модуль, чтобы сделать все значения положительными

print(data['dob_years'].max())
print(data['dob_years'].min())
(data[data['dob_years']==0])
```

Для удобства вставим вместо нулевых значений, медианные значения возраста

```python
def median_age(age):
    n_age = int(data['dob_years'].median())
    if age == 0:
        age = n_age
    return age

data['dob_years'] = data['dob_years'].apply(median_age)
print('Минимальный возраст заемщика:',data['dob_years'].min(), 'лет')
```

# Шаг 6. Категоризация данных.

```python
#Категоризируем данные о способе получения дохода.
def category(income):
    if income <= 30000:
        return 'E'
    if income <= 50000:
        return 'D'
    if income <= 200000:
        return 'C'
    if income <= 1000000:
        return 'B'
    return 'A'

data['total_income_category'] = data['total_income'].apply(category)
data['total_income_category'].value_counts()
```

```python
#Категоризируем данные по возрасту клиента.

def years_old(dob_years):
    if dob_years <= 25:
        return 'junior'
    if dob_years <= 35:
        return 'middle'
    if dob_years <= 50:
        return 'adult'
    return 'retiree'
data['age_group'] = data['dob_years'].apply(years_old)
data['age_group'].value_counts()
```

```python
credit_history['purpose'].value_counts()
```

```python
#Категоризируем данные по целям кредитов.

def purpose_of_the_loan(purpose):
    if 'свадьб' in purpose:
        return 'свадьба'
    if 'недвижим' in purpose:
        return 'недвижимость'
    if 'авто' in purpose:
        return 'автомобиль'
    if 'образ' in purpose:
        return 'образование'
    if 'жил' in purpose:
        return 'недвижимость'

data ['purpose_of_the_loan'] = data['purpose'].apply(purpose_of_the_loan)
data['purpose_of_the_loan'].value_counts()
```
